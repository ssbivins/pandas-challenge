```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
# Create dataframe from merged files
merged_df = pd.DataFrame(school_data_complete)
# Look at headings and first few rows
merged_df.head()
# Fix headings
renamed_df = merged_df.rename(columns={"student_name":"Student Name","gender":"Gender","grade":"Grade","school_name":"School Name","reading_score":"Reading Score","math_score":"Math Score","type":"Type","size":"# Students","budget":"Budget $"})
renamed_df.head()
# Get rid of unneeded columns
reduced_df = renamed_df[["Student ID","Student Name","Gender","Grade","School Name","Reading Score","Math Score","Type","# Students","Budget $"]]
# Don't show the index

# Show first few rows
reduced_df.set_index('Student ID', inplace=True)
reduced_df.head()
#Maybe the students with "Dr." in front of their names are real-life Doogie Howser, MDs

```

```python
## District Summary

# Remove rows with missing information
reduced_df = reduced_df.dropna(how='any')

# Calculate the total number of schools = 15
number_of_schools = reduced_df['School Name'].nunique()
f"Number of Schools = {number_of_schools}"

# Calculate the total number of students = 39170
number_of_students = reduced_df['Student Name'].count()
f"Number of Students = {number_of_students}"

# Calculate the total budget = Total Budget = $82932329558
total_budget = school_data['budget'].sum()
f"Total Budget = ${total_budget}"

# Calculate the average math score = 78.98537145774827 
average_math_score = reduced_df["Math Score"].mean()
f"Average Math Score = {average_math_score}"

# Calculate the average reading score = 81.87784018381414
average_reading_score = reduced_df["Reading Score"].mean()
f"Average Reading Score = {average_reading_score}"

# Calculate the percentage of students with a passing math score (70 or greater)
# Get the rows of students who passed math
df_passed_math = reduced_df.loc[reduced_df["Math Score"] >= 70, :]
# Get the number of students who passed math
num_passed_math = len(df_passed_math)
f"Number of students who passed math = {num_passed_math}"
#Calculate the percent of students who passed math
percent_passed_math = (num_passed_math / number_of_students * 100).round(2).astype(str) + "%"
f"Percent of students who passed math = {percent_passed_math}"

# Calculate the percentage of students with a passing reading score (70 or greater)
# Get the rows of students who passed reading
df_passed_reading = reduced_df.loc[reduced_df["Reading Score"] >= 70, :]
# Get the number of students who passed reading
num_passed_reading = len(df_passed_reading)
f"Number of students who passed reading = {num_passed_reading}"
#Calculate the percent of students who passed reading
percent_passed_reading = (num_passed_reading / number_of_students * 100).round(2).astype(str) + "%"
f"Percent of students who passed reading = {percent_passed_reading}"
df_passed_math.head()
df_passed_reading.head()

# Calculate the percentage of students who passed math **and** reading (% Overall Passing)
# Merge the two data frames using an inner join
df_passed_both = pd.merge(df_passed_math,df_passed_reading, on = "Student ID",how = "inner")
num_passed_both = len(df_passed_both)
percent_passed_both = (num_passed_both / number_of_students * 100).round(2).astype(str) + "%"
f"Percent of students who passed both Math and Reading = {percent_passed_both}"

# Create a dataframe to hold the above results
district_list = [{"Total Schools": number_of_schools,"Total Students": number_of_students,"Total Budget": total_budget,
                "Average Math Score": average_math_score,"Average Reading Score": average_reading_score,"% Passing Math": percent_passed_math,
                "% Passing Reading": percent_passed_reading,"% Overall Passing": percent_passed_both}]
district_df = pd.DataFrame(district_list)


# Optional: give the displayed data cleaner formatting
district_df["Total Students"] = district_df["Total Students"].map("{:,}".format)
district_df["Total Budget"] = district_df["Total Budget"].map("${:,}".format)
district_df["Average Math Score"] = district_df["Average Math Score"].map("{:.2f}".format)
district_df["Average Reading Score"] = district_df["Average Reading Score"].map("{:.2f}".format)
# district_df.set_index("Total Schools", inplace=True)
district_df.head()




```
